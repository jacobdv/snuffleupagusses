```python

import pandas as pd
```

```python
area_table_data = pd.read_csv("data/Area_Table_June_2020_V1.csv", usecols=["id", "speed", "has_1"])
area_table_df = pd.DataFrame(area_table_data)
area_table_df
```

```python
geography_data = pd.read_csv("data/Geography_Lookup_Table.csv")
geo_df = pd.DataFrame(geography_data)
geo_df
```

```python
area_table_df.rename(columns={"id": "geoid"}, inplace=True)
area_table_df
```

```python
area_table_df.drop_duplicates(inplace=True)
area_table_df
```

```python
internet = pd.merge(geo_df, area_table_df, how="inner", on="geoid")
internet
```

```python
internet_reduce = internet[["Year", "geoid", "name", "centroid_lng", "centroid_lat", "has_1"]]
internet_reduce
```

```python
internet2020 = internet_reduce[internet_reduce["Year"]== 2020]
internet2020.tail(22)
```

```python
internet2020.sort_values("has_1")
```

```python
internet2020.drop_duplicates(subset=["geoid"], inplace=True)
internet2020
```

```python
internet2020 = internet2020[["geoid", "name", "centroid_lng", "centroid_lat", "has_1"]]
internet2020
```

```python
internet2020.rename(columns={"centroid_lng": "longitude", "centroid_lat": "latitude", "has_1": "pop_with_access"}, inplace=True)
internet2020
```

```python
internet2020.to_csv("data/internet2020.csv", encoding="utf-8", index=False)
```
